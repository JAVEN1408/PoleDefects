```python
import cv2
import matplotlib.pyplot as plt
from inference_sdk import InferenceHTTPClient
```

```python

# Create an inference client
CLIENT = InferenceHTTPClient(
    api_url="https://detect.roboflow.com",
    api_key="YZnWJL3fpAiUh1u8a5mv"
)
```

```python
# Function to perform inference and display the image with predictions
def display_inference(image_path, model_id):
    # Perform inference
    result = CLIENT.infer(image_path, model_id=model_id)
    
    # Load the image
    img = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Parse the result to extract predictions
    predictions = result['predictions']
    
    # Display image and overlay predictions
    plt.imshow(img_rgb)
    ax = plt.gca()

    for pred in predictions:
        x1 = int(pred['x'] - pred['width'] / 2)
        y1 = int(pred['y'] - pred['height'] / 2)
        x2 = int(pred['x'] + pred['width'] / 2)
        y2 = int(pred['y'] + pred['height'] / 2)
        label = pred['class']
        conf = pred['confidence']
        
        # Draw bounding box
        rect = plt.Rectangle((x1, y1), x2 - x1, y2 - y1, fill=False, color='red', linewidth=2)
        ax.add_patch(rect)
        
        # Draw label and confidence
        plt.text(x1, y1 - 10, f'{label} {conf:.2f}', color='red', fontsize=12, bbox=dict(facecolor='yellow', alpha=0.5))

    plt.axis('off')
    plt.show()
```

```python
# Run the function
display_inference(image_path, model_id="c-mwzd6/1")
```
